```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [i+1 for i in range(10)]

y1 = np.asarray([0.8672, 0.9223, 0.9396, 0.9498, 0.9545, 0.9594, 0.9627, 0.9654, 0.9681, 0.9692])
y2 = np.asarray([0.7744, 0.1452, 0.936, 0.7816, 0.9617, 0.9634, 0.9668, 0.9582, 0.2287, 0.9684])
y3 = np.asarray([0.8672, 0.9223, 0.9396, 0.9464, 0.9535, 0.9331, 0.9624, 0.9656, 0.9452, 0.9687])



# make some data
fig, ax = plt.subplots(figsize=(10, 6), dpi= 100, facecolor='w', edgecolor='k')

ax.plot(x, y1*100, linewidth=2.5, label='Normal training', marker='*')
ax.plot(x, y2*100, linewidth=2.5, label='Untargeted attack training', marker='o')
ax.plot(x, y3*100, linewidth=2.5, label='Targeted attack training', marker='>')

leg = ax.legend(ncol=1, shadow=True, title="Training Type", fancybox=True)

for line in leg.get_lines():
    line.set_linewidth(4.0)

leg.get_title().set_color("red")

plt.xlabel("Global Round")
plt.ylabel("Accuracy %")
plt.title("Simple Environment\n Global Accuracy")
plt.xticks(np.arange(1, 10, step=1))
plt.yticks(np.arange(0, 100, step=10))
ax.grid(True)
plt.show()
```
